```python
import numpy as np 
import pandas as pd 
import random as rn
import re
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import seaborn as sns
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'   #파이썬 텐서플로에서 ERROR, INFO, WARNING 로그를 출력하지 않는 방법
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
```

```python
rn.seed(a=42)
p = 0.005

df_reviews = pd.read_csv('/content/drive/MyDrive/Section4/dataset.csv',
                         skiprows=lambda i: i>0 and rn.random() > p)

print(df_reviews.shape)
display(df_reviews.head())
```

```python
# string으로 변경
df_reviews["review_text"] = df_reviews["review_text"].astype(str)
df_reviews["review_votes"] = df_reviews["review_votes"].astype(str)
df_reviews.review_text = df_reviews.review_text.apply(lambda s: s.strip())

# null 제거
df_reviews_2 = df_reviews[df_reviews["review_score"].notnull()]

# 1, -1 를 1 과 0 으로변경
df_reviews_2["review_score"] = \
np.where(df_reviews_2["review_score"]==-1, 0, df_reviews_2["review_score"])
```

```python
# "Early Access Review" comments 지우기
df_reviews_2 = df_reviews_2[df_reviews_2.review_text != "Early Access Review"]
df_reviews_2 = df_reviews_2[~df_reviews_2.review_text.isin(['nan'])]
print(df_reviews_2.shape)

# Drop duplicates 
df_reviews_2.drop_duplicates(['review_text', 'review_score'], inplace = True)
print(df_reviews_2.shape)
```

```python
df_reviews_2['word_count'] = df_reviews_2.review_text.apply(lambda s: len(s.split()))

print(df_reviews_2.word_count.min())
print(df_reviews_2.word_count.quantile(0.05))
print(df_reviews_2.word_count.quantile(0.25))
print(df_reviews_2.word_count.median())
print(df_reviews_2.word_count.quantile(0.95))
print(df_reviews_2.word_count.quantile(0.99))
print(df_reviews_2.word_count.max())
```

```python
a = sns.histplot(data=df_reviews_2, x="word_count", binwidth=40)
```

```python
## Text Cleaning
def replace_hearts_with_PAD(text):
    return re.sub(r"[♥]+", ' **** ' ,text)

df_reviews_2['review_text_clean'] = df_reviews_2.review_text.apply(replace_hearts_with_PAD)
```

```python
df_reviews_3 = df_reviews_2[['review_text_clean', 'review_score']]
df_reviews_3 = df_reviews_3.rename({"review_text_clean": "text", "review_score": "labels"});
df_reviews_3.head()
```

```python
# 데이터를 스플릿 해주었습니다.(60-20-20)

train_df, eval_df = train_test_split(df_reviews_3, test_size = 0.4, random_state = 42)
test_df , holdout_df = train_test_split(eval_df, test_size = 0.5, random_state = 42)

print(train_df.shape)
print(test_df.shape)
print(holdout_df.shape)
```

```python
from simpletransformers.classification import ClassificationModel
import logging

logging.basicConfig(level=logging.INFO)
transformers_logger = logging.getLogger("transformers")
transformers_logger.setLevel(logging.WARNING)


# Create a ClassificationModel
model_type = "roberta"
model_name = "roberta-base"

# roberta는 bert의 변형이다.
# BERT : 'B'idirectional 'E'ncoder 'R'epresentations from 'T'ransformersPermalink
# BERT: Pre-training of Deep Bidirectional Transformers for Language Understanding 논문에 기록되어 있습니다.
# 18년 10월 공개한 구글의 새로운 language representation model
# NLP 11개의 task에서 최고 성능을 보임

# roberta 와 bert의 차이 
#1.큰 형태의 말뭉치 데이터
#2.동적 데이터 마스킹 -> DDM(동적 데이터 마스킹)에서는 권한이 없는 사용자로 마스킹하여 중요한 데이터 노출을 제한합니다. 애플리케이션의 보안 설계 및 코딩을 크게 간소화하는 데 사용됩니다.
#  동적 데이터 마스킹의 목적은 중요한 데이터의 노출을 제한하여 데이터에 대한 액세스 권한이 없는 사용자가 보지 못하게 하는 것이지
#  데이터베이스 사용자가 데이터베이스에 직접 연결하여 중요한 데이터 조각을 노출하는 쿼리를 억지로 실행하지 못하게 하는 것은 아닙니다. 

# 구글링을 통해서 로버타 모델을 불러오는 것을 참고해 작성해보았습니다.
roberta_model = ClassificationModel(
                          model_type,
                          model_name,
                          args={'num_train_epochs' : 1,
                                 "train_batch_size": 16,
                                 "eval_batch_size": 16,
                                 "fp16": False,
                                 "optimizer": "AdamW",
                                 "adam_epsilon": 1e-8,
                                 "learning_rate": 1e-5,
                                 "weight_decay": 0.7,
                                 'overwrite_output_dir': True,
                                 "save_eval_checkpoints": False,
                                 "save_model_every_epoch": False,
                                 "no_cache": True,
                                 "manual_seed": 12345})


#num_train_epochs를 1로 설정하고 다음 루프에서 훈련을 반복하는 이유는
#트랜스포머 모델에서 학습률은 에폭/(총 에폭 수)의 영향을 받기 때문입니다.
#가장 좋은 모델이 네번째 모델이었다고 칠 때
#num_train_steps =4로 재실행하면 num_train_trains=5의 네번째 에폭에서 봤던 결과랑 다르게 나온다고 합니다.
#하지만 이렇게 훈련을 반복하면 그런 문제가 발생하지 않는다는 정보를 얻었습니다.


for i in range(2):

    # Train the model
    roberta_model.train_model(train_df)
    # Evaluate the model
    result, model_outputs, wrong_predictions = roberta_model.eval_model(test_df)
    print("Accuracy= " ,(result['tp'] + result['tn']) / (result['tp'] + result['tn'] + \
                                                         result['fp'] + result['fn']))
    print("Recall = ",(result['tn']) / (result['tn'] + result['fn'])) # simpletransformers fn과 fp를 잘못 출력하기 떄문에 바꾸어준다.
    print(result)
    print(classification_report(np.argmax(model_outputs, axis = 1), test_df.review_score.values))
```
